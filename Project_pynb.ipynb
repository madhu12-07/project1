```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("IBM HR Employee Attrition Data.csv")
print(df.head())
print(df.info())
```

```python
print(df.isnull().sum())
```

```python
print(df.columns)
```

```python
df = df.drop(["EmployeeCount","Over18","StandardHours"], axis=1)
```

```python
sns.countplot(x="Attrition", data=df, palette="Set2")
plt.title("Attrition Count")
plt.show()
```

```python
plt.figure(figsize=(8,5))
sns.countplot(x="Department", hue="Attrition", data=df, palette="Set2")
plt.title("Attrition by Department")
plt.show()
```

```python
plt.figure(figsize=(8,5))
sns.boxplot(x="Attrition", y="MonthlyIncome", data=df, palette="coolwarm")
plt.title("Attrition vs Monthly Income")
plt.show()
```

```python
plt.figure(figsize=(10,6))
sns.countplot(y="JobRole", hue="Attrition", data=df, palette="viridis")
plt.title("Attrition by Job Role")
plt.show()
```

```python
plt.figure(figsize=(12,8))
sns.heatmap(df.select_dtypes(include="number").corr(),
            annot=True, cmap="Blues", annot_kws={"size":5.8})
plt.title("Correlation Heatmap (Numeric Features)")
plt.show()
```

```python
plt.figure(figsize=(12,8))
sns.heatmap(df.select_dtypes(include="number").corr().round(2),
            annot=True, cmap="Blues", annot_kws={"size":12})
plt.title("Correlation Heatmap (Rounded Values)")
plt.show()
```

```python
corr = df.select_dtypes(include="number").corr()
plt.figure(figsize=(12,8))
sns.heatmap(corr[(corr>0.3) | (corr<-0.3)],
            annot=True, cmap="Blues", annot_kws={"size":12})
plt.title("Heatmap (Strong Correlations Only)")
plt.show()
```
